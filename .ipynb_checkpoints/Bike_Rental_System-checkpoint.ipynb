### About Dataset

This dataset has the following fields:

- instant    : record index
- dteday     : date
- season     : season (1:springer, 2:summer, 3:fall, 4:winter)
- yr         : year (0: 2011, 1:2012)
- mnth       : month ( 1 to 12)
- hr         : hour (0 to 23)
- holiday    : weather day is holiday or not (extracted from [Web Link])
- weekday    : day of the week
- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
- weathersit :  
                - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
                - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
                - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
                - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp       : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
- atemp      : Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
- hum        : Normalized humidity. The values are divided to 100 (max)
- windspeed  : Normalized wind speed. The values are divided to 67 (max)
- casual     : count of casual users
- registered : count of registered users
- cnt        : count of total rental bikes including both casual and registered

```python
# Removing outliers - data points that are 
bike_df_wo = bike_df[np.abs(bike_df["cnt"]-bike_df["cnt"].mean())<=(3*bike_df["cnt"].std())] 
```

```python
print ("Dataset shape before removing outliers: ",bike_df.shape)
print ("Dataset shape after removing the outliers: ",bike_df_wo.shape)
```

### Dummy variables
Creating dummy variables to include categorical variables such as weather, season, month, etc. in our model.

```python
dummy_features = ['season', 'weathersit', 'mnth', 'hr', 'weekday']
for each in dummy_features:
    dummies = pd.get_dummies(bike_df_wo[each], prefix=each, drop_first=True)
    bike_df_wo = pd.concat([bike_df_wo, dummies], axis=1)

drop_features = ['instant', 'dteday', 'season', 'weathersit', 'weekday', 'casual', 'registered', 'atemp', 'mnth', 'workingday', 'hr']
ride_df = bike_df_wo.drop(drop_features, axis=1)
ride_df.head()
```

### Scaling continuous variables
Standardizing each of the continuous variables i.e. we'll shift and scale the variables such that they have zero mean and a standard deviation of 1.

The scaling factors are saved so we can go backwards when we use the network for predictions.

```python
quant_features = ['cnt', 'temp', 'hum', 'windspeed']
# Store scalings in a dictionary so we can convert back later
scaled_features = {}
for each in quant_features:
    mean, std = ride_df[each].mean(), ride_df[each].std()
    scaled_features[each] = [mean, std]
    ride_df.loc[:, each] = (ride_df[each] - mean)/std
```

### Splitting the data into training, testing, and validation sets
We'll save the data for the last approximately 30 days to use as a test set. 
Splitting the train set into train data and validation data to validate the model using the last 60 days of the train set. 
We'll use train set to make predictions and compare them with the actual number of riders.

```python
# Save data for approximately the last 21 days 
test_df = ride_df[-30*24:]

# Now remove the test data from the data set 
ride_df = ride_df[:-30*24]

# Separate the data into features and targets
target_field = ['cnt']
features, target = ride_df.drop(target_field, axis=1), ride_df[target_field]
X_test, y_test = test_df.drop(target_field, axis=1).values, test_df[target_field].values.ravel()

# Hold out the last 60 days or so of the remaining data as a validation set
X_train, y_train = features[:-60*24].values, target[:-60*24].values.ravel()
X_valid, y_valid = features[-60*24:].values, target[-60*24:].values.ravel()
```

### Model Selection

```python
table = PrettyTable()
table.field_names = ["Model", "Mean Absolute Error" ,"Root Mean Squared Error", "R² score"]

models = [
    SGDRegressor(max_iter=2000, tol=1e-3),
    Lasso(alpha=0.1),
    Ridge(alpha=.5),
    SVR(gamma='auto', kernel='linear'),
    SVR(gamma='auto', kernel='rbf'),
    RandomForestRegressor( random_state=0, n_estimators=100)
]

for model in models:
    model.fit(X_train, y_train) 
    y_res = model.predict(X_valid)

    mae = mean_absolute_error(y_valid, y_res)
    rmse = np.sqrt(mean_squared_error(y_valid, y_res))
    score = model.score(X_valid, y_valid) 
    table.add_row([type(model).__name__, format(mae, '.2f'), format(rmse, '.2f'), format(score, '.2f')])

print(table)
```

#### We can see that the Random Forest is performing the best among other models

```python
# Defining function to display the model metrics
def evaluate(x, y, dataset):
    pred = model.predict(x)
    mae = mean_absolute_error(y, pred)
    rmse = mean_squared_error(y, pred)
    score = model.score(x, y)    

    table.add_row([type(model).__name__, dataset, format(mae, '.2f'), format(rmse, '.2f'), format(score, '.2f')])
```

### Performing Grid Search to fine tuning the selected model

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 200, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(2, 20, num = 10)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)

# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
model_rs = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, cv = 10, verbose=2, random_state=0)
# Fit the random search model
model_rs.fit(X_train, y_train)
print(model_rs)

# Table setup
table = PrettyTable()
table.field_names = ["Model", "Dataset", "MAE", 'RMSE', "R² score"]

evaluate(X_train, y_train, 'training')
evaluate(X_valid, y_valid, 'validation')
evaluate(X_test, y_test, 'test')

print(table)
```
